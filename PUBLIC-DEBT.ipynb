```python
import matplotlib.pyplot as plt
import seaborn as sns

# A Barplot of CREDITOR NAME against AMOUNT OUTSTANDING AS AT 30/06/2023\n(Ksh Equivalent)
plt.figure(figsize=(15,10))
plt.bar(data['CREDITOR NAME'],data['AMOUNT OUTSTANDING AS AT 30/06/2023\n(Ksh Equivalent)'],color='red')
plt.title('Outstanding Debt as at 30/06/2023')
plt.xlabel('CREDITOR')
plt.ylabel('AMOUNT OUTSTANDING(KES)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data['AMOUNT OUTSTANDING AS AT 30/06/2023\n(Ksh Equivalent)'], kde=True)
plt.title('Distribution of Outstanding Debt Amount')
plt.xlabel('Outstanding Amount')
plt.ylabel('Frequency')
plt.show()
```
